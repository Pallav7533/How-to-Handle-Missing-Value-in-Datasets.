# Data Standardization

The Process of standardization the data to a comman format and a comman range.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import sklearn.datasets
```

```python
data = sklearn.datasets.load_breast_cancer()
```

```python
data
```

```python
#create a dataframe

df1 = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df1['target'] = data.target
```

```python
df1.head()
```

```python
df1.shape
```

```python
x = df1
y = data.target
```

```python
print(x)
```

```python
print(y)
```

```python
X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state = 2)
```

```python
print(x.shape , X_train.shape , X_test.shape)
```

```python
print(data.data.std())
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
x_train_stdd = scaler.transform(X_train)
```

```python
print(x_train_stdd.std())
```

```python
print(x_train_stdd)
```

# LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
#load the label encoder function

label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(data.diagnosis)
```

```python
data['target'] = labels
```

```python
data.head()
```

```python
data['target'].value_counts()
```
